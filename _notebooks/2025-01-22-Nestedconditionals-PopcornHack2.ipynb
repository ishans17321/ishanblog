### Popcorn Hack 2

This code uses boolean to check if a person has the unlimited amusement food card. If they do, then they are allowed to buy any food that they want for free. Add code for another type of food and take into account whether or not the food is in stock.

```python
%%html
<!-- HTML output div -->
<div id="message"></div>
<script>
    function runWhenDOMLoaded() {
        // Determines if the user has the food card
        let foodCard = true; 

        // Amount of money the user has
        let money = 14; 

        // Food choice made by the user
        let foodChoice = "pretzels"; 

        // Availability of foods
        let popcornInStock = false; 
        let cottonCandyInStock = true; 
        let pretzelsInStock = true; 

        // Get the HTML element to display messages
        const messageElement = document.getElementById("message");

        /**
         * Function to display messages on the webpage and log them to the console
         * @param {string} message - The message to display
         */
        function displayMessage(message) {
            console.log(message); // Log the message to the console
            messageElement.textContent += message + "\n"; // Append the message to the webpage
        }

        // Display the current stock status
        displayMessage("Items currently in stock:");
        if (popcornInStock) displayMessage("- Popcorn");
        if (cottonCandyInStock) displayMessage("- Cotton Candy");
        if (pretzelsInStock) displayMessage("- Pretzels");
        displayMessage("\n"); // Add a blank line for better readability

        // Check if the user has the food card
        if (foodCard === true) {
            // The user can get any food for free
            if (foodChoice === "popcorn") {
                if (popcornInStock) {
                    displayMessage("Yum! The popcorn is tasty! (8 dollars with your food card)");
                } else {
                    displayMessage("Sorry, popcorn is out of stock.");
                }
            } else if (foodChoice === "cotton candy") {
                if (cottonCandyInStock) {
                    displayMessage("Yum! The cotton candy is tasty! (5 dollars with your food card)");
                } else {
                    displayMessage("Sorry, cotton candy is out of stock.");
                }
            } else if (foodChoice === "pretzels") {
                if (pretzelsInStock) {
                    displayMessage("Yum! The pretzels are tasty! (6 dollars with your food card)");
                } else {
                    displayMessage("Sorry, pretzels are out of stock.");
                }
            } else {
                displayMessage("Sorry, we don't have that food.");
            }
        } else {
            // If the user doesn't have the food card, check if they have enough money
            if (foodChoice === "popcorn") {
                if (money >= 8) { // Popcorn costs $8
                    if (popcornInStock) {
                        displayMessage("Yay! You can buy the popcorn!");
                    } else {
                        displayMessage("Sorry, popcorn is out of stock.");
                    }
                } else {
                    displayMessage("Oh... you can't buy the popcorn (not enough money).");
                }
            } else if (foodChoice === "cotton candy") {
                if (money >= 5) { // Cotton candy costs $5
                    if (cottonCandyInStock) {
                        displayMessage("Yay! You can buy the cotton candy!");
                    } else {
                        displayMessage("Sorry, cotton candy is out of stock.");
                    }
                } else {
                    displayMessage("Oh... you can't buy the cotton candy (not enough money).");
                }
            } else if (foodChoice === "pretzels") {
                if (money >= 6) { // Pretzels cost $6
                    if (pretzelsInStock) {
                        displayMessage("Yay! You can buy the pretzels!");
                    } else {
                        displayMessage("Sorry, pretzels are out of stock.");
                    }
                } else {
                    displayMessage("Oh... you can't buy the pretzels (not enough money).");
                }
            } else {
                displayMessage("Sorry, we don't have that food.");
            }
        }
    }

    // Ensure the function runs only after the page loads
    if (document.readyState === "loading") {
        document.addEventListener("DOMContentLoaded", runWhenDOMLoaded);
    } else {
        runWhenDOMLoaded();
    }
</script>
```
